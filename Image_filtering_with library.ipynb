# FIND EDGES,HORIZANTAL EDGES,AND VERTICAL EDGES BY APPLYING FILTERS

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy.signal import convolve2d
import urllib.request
```

```python
# bb=cv2.imread('/content/drive/MyDrive/PLC/top-view-wooden-frame-with-copy-space.jpg')
# bb=cv2.imread('/content/drive/MyDrive/PLC/bedroom-interior-living-area-nature-mockup-farmhouse-style-3d-rendering.jpg')
url = "https://image.freepik.com/free-photo/bedroom-interior-living-area-nature-mockup-farmhouse-style-3d-rendering_41470-4069.jpg"
url_response = urllib.request.urlopen(url)
img_array = np.array(bytearray(url_response.read()), dtype=np.uint8)
bb = cv2.imdecode(img_array, -1)
bb2=cv2.cvtColor(bb,cv2.COLOR_BGR2GRAY)
bbrgb=cv2.cvtColor(bb,cv2.COLOR_BGR2RGB)
plt.imshow(bb2,cmap='gray')
plt.show()
```

```python
# vertical sobel filter detect the vertical lines or edges
v_fil = np.array([[1, 0, -1],
                    [2, 0, -2],
                    [1, 0, -1]])
# Horizantal sobel filter detect the horizantal lines or edges
h_fil=np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])
# And to detect all edges use this filter
kernel_edgedetection = np.array([[-1, -1, -1],
                                 [-1, 8.5, -1],
                                 [-1, -1, -1]])
```

```python

Hori = convolve2d(bb2, h_fil)
Ver=convolve2d(bb2, v_fil)
Edges=convolve2d(bb2, kernel_edgedetection)

plt.figure(figsize=(20,5))

plt.subplot(1,4,1)
plt.imshow(bbrgb)
plt.title('Original Photo')
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(abs(Edges) , cmap='gray')
plt.title('Edges find Filter')
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(abs(Hori) , cmap='gray')
plt.title('Horizontal Sobel Filter')
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(abs(Ver) , cmap='gray')
plt.title('Vertical Sobel Filter')
plt.axis("off")

plt.show()
```
